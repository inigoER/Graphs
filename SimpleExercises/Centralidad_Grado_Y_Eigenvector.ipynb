```python
pos = nx.spring_layout(G)
fig = plt.figure(1, figsize=(20,8), dpi=100)
nx.draw(G,pos, with_labels=True)
plt.title("Game of thrones betrayals network")
plt.show()
```

```python
import collections
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=0.80, color='b')

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax.set_xticks([d for d in deg])
ax.set_xticklabels(deg)

plt.show()
```

```python
GDegCent = nx.degree_centrality(G)
GEigCent = nx.eigenvector_centrality(G)
```

```python
sorted((v, f"{c:0.2f}") for v, c in GDegCent.items())
```

```python
sorted((v, f"{c:0.2f}") for v, c in GEigCent.items())
```

Los datos son un grafo dirigido en el que los nodos son personajes de la serie Game of Thrones y las aristas son dirigidas indicando si un personaje ha traicionado a otro. 

Hay 17 nodos, para 17 personajes con 39 aristas, o 39 traiciones.

El histograma de grados parece indicar que la mayoría de los nodos han realizado o recibido menos de 3 traiciones, pero hay algunos que han recibido o realizado más.

La centralidad de grado asigna a Cersei a Petyr, a Tyrion, a Tywin y Varys (en ese orden) como los vértices más centrales, esto tiene sentido, Cersei es la persona que más traiciones realiza y recibe, Petyr y Varys son los personajes de la serie que "manejan" o "causan" casi todos los eventos desde las sombras, Tyrion y Tywin son los dos lannister más inteligentes y que más planes realizan.

La centralidad de eigenvector asigna a Cersei, Jaime, Tywin y Tyrion como los vértices más centrales, estos son los lannister de la serie, los cuales podría decirse son la familia más involucrada en traiciones y planes secretos.

# Grafo internet
Este grafo es "Zachary's karate club: social network of friendships between 34 members of a karate club at a US university in the 1970s. Please cite W. W. Zachary, An information flow model for conflict and fission in small groups, Journal of Anthropological Research 33, 452-473 (1977)." del link http://www-personal.umich.edu/~mejn/netdata/

```python
G2 = nx.read_gml('karate.gml', label = None)
```

```python
pos = nx.spring_layout(G2)
fig = plt.figure(1, figsize=(20,8), dpi=100)
nx.draw(G2,pos, with_labels=True)
plt.title("Karate Friends network")
plt.show()
```

```python
import collections
degree_sequence = sorted([d for n, d in G2.degree()], reverse=True)  # degree sequence
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=0.80, color='b')

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax.set_xticks([d for d in deg])
ax.set_xticklabels(deg)

plt.show()
```

```python
G2.number_of_edges()
```

```python
G2DegCent = nx.degree_centrality(G2)
G2EigCent = nx.eigenvector_centrality(G2)
```

```python
sorted((v, f"{c:0.2f}") for v, c in G2DegCent.items())
```

```python
sorted((v, f"{c:0.2f}") for v, c in G2EigCent.items())
```

Los datos son un grafo no dirigido en el que los nodos son personas que van a karate y las aristas indican si dos personas son amigas.

Hay 34 nodos, para 34 personas con 78 aristas, o 78 amistades.

El histograma de grados parece indicar que la mayoría de los nodos tienen 2 amigos, específicamente entre 1 y 6, pero hay algunos que llegan a tener hasta 17.

La centralidad de grado asigna a los vértices 34, 1, 33, 3 y 2 (en ese orden) como los vértices más centrales, en el grafo, podemos ver que estos vértices sí tienen muchas conexiones a otros vértices.

La centralidad de eigenvector asigna al 34, 1, 33, 3 y 2 de nuevo (también en ese orden), la similitud en ambos métodos probablemente se deba a que estos vértices están conectados a varios y también están conectados a otros vértices importantes.
