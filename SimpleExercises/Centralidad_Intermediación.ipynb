```python
pos = nx.spring_layout(G)
fig = plt.figure(1, figsize=(20,8), dpi=100)
nx.draw(G,pos, with_labels=True)
plt.title("Game of thrones betrayals network")
plt.show()
```

```python
GIntCent = nx.betweenness_centrality(G)
```

```python
sorted((v, f"{c:0.2f}") for v, c in GIntCent.items())
```

Los datos son un grafo dirigido en el que los nodos son personajes de la serie Game of Thrones y las aristas son dirigidas indicando si un personaje ha traicionado a otro. 

Hay 17 nodos, para 17 personajes con 39 aristas, o 39 traiciones.

La centralidad de intermediacin asigna a Cersei, a Petyr, a Tyrion, a Varys y a Joffrey como los vrtices ms centrales, estos personajes son los ms influyentes en la serie, Cersei es la reina madre y tiene muchos espas, Petyr es una de las personas detrs de casi todos los eventos de la serie igual que Varys, Tyrion es el Lannister ms inteligente y el segundo ms influyente, Joffrey es el rey, esta centralidad parece haber obtenido a los personajes ms influyentes en la red de traiciones de la manera correcta.

# Grafo internet
Este grafo es "Zachary's karate club: social network of friendships between 34 members of a karate club at a US university in the 1970s. Please cite W. W. Zachary, An information flow model for conflict and fission in small groups, Journal of Anthropological Research 33, 452-473 (1977)." del link http://www-personal.umich.edu/~mejn/netdata/

```python
G2 = nx.read_gml('karate.gml', label = None)
```

```python
pos = nx.spring_layout(G2)
fig = plt.figure(1, figsize=(20,8), dpi=100)
nx.draw(G2,pos, with_labels=True)
plt.title("Karate Friends network")
plt.show()
```

```python
G2.number_of_edges()
```

```python
G2IntCent = nx.betweenness_centrality(G2)
```

```python
sorted((v, f"{c:0.2f}") for v, c in G2IntCent.items())
```

Los datos son un grafo no dirigido en el que los nodos son personas que van a karate y las aristas indican si dos personas son amigas.

Hay 34 nodos, para 34 personas con 78 aristas, o 78 amistades.

La centralidad de Katz asigna a los vrtices 1, 34, 33, 32 y 3 (en ese orden) como los vrtices ms centrales, esta es la primera vez que sale un resultado diferente de los vrtices ms centrales, sin embargo, siguen estando el 1 y el 34 en las primeras posiciones.
