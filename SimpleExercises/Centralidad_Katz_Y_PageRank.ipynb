```python
pos = nx.spring_layout(G)
fig = plt.figure(1, figsize=(20,8), dpi=100)
nx.draw(G,pos, with_labels=True)
plt.title("Game of thrones betrayals network")
plt.show()
```

```python
GKatzCent = nx.katz_centrality(G)
PageRank = nx.pagerank(G)
```

```python
sorted((v, f"{c:0.2f}") for v, c in GKatzCent.items())
```

```python
sorted((v, f"{c:0.2f}") for v, c in PageRank.items())
```

Los datos son un grafo dirigido en el que los nodos son personajes de la serie Game of Thrones y las aristas son dirigidas indicando si un personaje ha traicionado a otro. 

Hay 17 nodos, para 17 personajes con 39 aristas, o 39 traiciones.

La centralidad de Katz asigna a Cersei, a Tyrion, a Tywin, a Jaime y a Joffrey como los vértices más centrales, estos personajes son los Lannister, los personajes más involucrados en las traiciones, a diferencia de las otras centralidades, esta sí incluye a Joffrey, las demás solían dejarlo fuera.

Pagerank asigna a Cersei, Jaime, Tywin y Tyrion como los vértices más centrales, estos son los lannister de la serie sin incluir a Joffrey.

# Grafo internet
Este grafo es "Zachary's karate club: social network of friendships between 34 members of a karate club at a US university in the 1970s. Please cite W. W. Zachary, An information flow model for conflict and fission in small groups, Journal of Anthropological Research 33, 452-473 (1977)." del link http://www-personal.umich.edu/~mejn/netdata/

```python
G2 = nx.read_gml('karate.gml', label = None)
```

```python
pos = nx.spring_layout(G2)
fig = plt.figure(1, figsize=(20,8), dpi=100)
nx.draw(G2,pos, with_labels=True)
plt.title("Karate Friends network")
plt.show()
```

```python
G2.number_of_edges()
```

```python
G2KatzCent = nx.katz_centrality(G2)
PageRank2 = nx.pagerank(G2)
```

```python
sorted((v, f"{c:0.2f}") for v, c in G2KatzCent.items())
```

```python
sorted((v, f"{c:0.2f}") for v, c in PageRank2.items())
```

Los datos son un grafo no dirigido en el que los nodos son personas que van a karate y las aristas indican si dos personas son amigas.

Hay 34 nodos, para 34 personas con 78 aristas, o 78 amistades.

La centralidad de Katz asigna a los vértices 34, 1, 33, 3 y 2 (en ese orden) como los vértices más centrales, en el grafo, podemos ver que estos vértices sí tienen muchas conexiones a otros vértices, este orden de centralidad es el mismo que se obtiene en la centralidad de grado y eigenvector.

PageRank también asigna en ese orden los más centrales, con el 34 y 1 empatados, debido a que en todas las medidas esos vértices han sido el top 5 y en el mismo orden, podría decirse que realmente son muy influyentes en el grafo.
