```python
import networkx as nx
import numpy as np
import random
import matplotlib.pyplot as plt
```

# Gr√°fica aleatoria

```python
def random_graph(nv,ne):
  G = nx.Graph()
  pos_e = []
  for i in range(nv):
    for j in range(nv):
      pos_e.append([str(i+1),str(j+1)])
  chosen_edges = random.sample(pos_e, ne)
  for i in chosen_edges:
    G.add_edge(i[0],i[1])
  return G
```

```python
G1 = random_graph(300,1000)
```

```python
pos = nx.spring_layout(G1)
fig = plt.figure(1, figsize=(20,8), dpi=100)
nx.draw(G1,pos, with_labels=True)
plt.title("Random graph")
plt.show()
```

```python
import collections
degree_sequence = sorted([d for n, d in G1.degree()], reverse=True)  # degree sequence
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=0.80, color='b')

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax.set_xticks([d for d in deg])
ax.set_xticklabels(deg)

plt.show()
```

Como podemos ver, el grado promedio efectivamente es 2*(1000)/300 = 6.6, aprox 6.

Este m√©todo de gr√°ficas aleatorias se usan en el modelado de redes sociales, epidemias, estructuras del internet, etc.

Podemos ver que el grado promedio es efectivamente (n-1)*p = (299)(0.015) = 4.5, aprox 5
