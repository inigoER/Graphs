```python
import networkx as nx
import numpy as np
import random
import matplotlib.pyplot as plt
```

# Gr√°fica aleatoria

```python
def price_pa_graph(n, m, alpha):
  G = nx.DiGraph()
  G.add_edge(0, 1)
  for i in range(2, n):
    # choose m nodes to connect to
    nodes = []
    while len(nodes) < m:
      node = random.choice(list(G.nodes()))
      prob = G.in_degree(node)**alpha
      if random.random() < prob:
        nodes.append(node)
      # connect node i to the selected nodes
    for node in nodes:
      G.add_edge(node, i)
  return G
```

```python
G1 = price_pa_graph(300, 2, 3)
```

```python
pos = nx.spring_layout(G1)
fig = plt.figure(1, figsize=(20,8), dpi=100)
nx.draw(G1,pos, with_labels=True)
plt.title("Random graph")
plt.show()
```

```python
import collections
degree_sequence = sorted([d for n, d in G1.degree()], reverse=True)  # degree sequence
# print "Degree sequence", degree_sequence
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())

fig, ax = plt.subplots()
plt.bar(deg, cnt, width=0.80, color='b')

plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax.set_xticks([d for d in deg])
ax.set_xticklabels(deg)

plt.show()
```

Como podemos ver, el grado promedio efectivamente es 2, o c.

Este m√©todo de gr√°ficas aleatorias se usa principalmente como vimos en clase en modelado de citaciones.
